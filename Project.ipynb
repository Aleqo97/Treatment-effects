**Генерируем контрольную переменную возраст $\text{Age}_i$**

как нормальное $\text{N}\left(45, 10^2\right)$

```python
# age

# Предположим, что распределение опыта работы схоже с нормальным
age = norm.rvs(size = n, loc = 45, scale = 10)

# Для красоты ограничим минимальные и максимальные значения
age[age >= 75] = 75
age[age <= 18]  = 18

# Также, для удобства введем округление
age = np.round(age)

# Посмотрим на несколько первых значений
print(age[0:10])
```

Генерируем переменную наличие работы, как $\text{Job}_{i}$ как распределение Бернулли

```python
# Сгенерируем переменную на факт проживания в городе
job = np.random.binomial(n = 1, p = 0.7, size = n)

# Доля индивидов, у которых есть работа
print(f'{np.round(np.mean(job)*100, 2)}%')
```

**Генерируем переменную наличие детей**

$\text{Children}_{i}\sim\text{Ber}\left(0.8\right)$.

```python
# Сгенерируем переменную на факт проживания в городе
children = np.random.binomial(n = 1, p = 0.8, size = n)

# Доля индивидов, у которых есть работа
print(f'{np.round(np.mean(children)*100, 2)}%')
```

**Генерируем инструментальную переменную**

$$\text{P}\left(\text{Block}_{i} = 1|\text{Age}_{i}, \text{Job}_{i}, \text{Children}_{i}\right) = \Phi\left(\underbrace{\frac{0.5 * \text{Children}_{i}}{ (1-\text{Job}_{i}- \text{Age}_{i})} + \text{Job}_{i}\times\text{Age}_{i}^{1/5}}_{\text{индекс}}\right)$$


```python
# Сформируем индекс
block_index = 2* (children +job)/ (1 + age) - 1.1*children**3*age**(1/4) + children*job
# Создадим условные вероятности
block_prob = norm.cdf(block_index)


# Сгенерируем Наличие у пользователя блокировщика рекламы на телефоне
block = np.random.binomial(n = 1, p = block_prob, size = n)

# Посмотрим на несколько первых условных вероятностей
print(block_prob[0:10])
```

```python
# Дисперсия индекса
print(np.var(block_index))
```

```python
# Доля пользователей с блокировщиком рекламы на телефоне
print(np.mean(block))
```

**Генерация ненаблюдаемой переменной, порождающей эндогенность**

**Мера интереса к видео.**  Логика: если видео очень интересное то рекламы ты посмотришь, так как нужно видео смотреть. Ну и очевидно если видео очень интересное то и само видео ты больше дольше смотреть.

```python
from scipy. stats import expon

interest = t.rvs(size = n, df = 10)
interest = (interest) * 10 + 50

interest[interest >= 100] = 100
interest[interest <= 1]   = 0
```

```python
# Посмотрим на распределение способностей
seaborn.histplot(interest,                   # данные
                 stat  = 'density',          # тип гистограммы
                 color = "seagreen",         # цвет гистограммы
                 label = "histogram",        # наименование гистограммы
                 bins  = 70)                 # число столбиков в Гистограмме
```

**Генерация переменной воздействия**

```python
# Равномерные случайные величины
u = uniform.rvs(size = n)
```

```python
# Сгенерируем часть индекса, не зависящую от наличия блокировщика
advert_index = interest/10 + 10*np.sqrt(job) + children *age - interest*children*1.1
```

```python
# Симулируем факт просмотра рекламы, когда у человека есть блокировщик

block1  = 1

advert1_index = advert_index - 40 * block1
advert1_prob  = logistic.cdf(advert1_index, scale = 1)
advert1       = (advert1_prob >= u).astype(int)

np.mean(advert1)
```

```python
# Симулируем факт просмотра рекламы, когда у человека нет блокировщика

block0    = 0
advert0_index = advert_index - 40 * block0
advert0_prob  = logistic.cdf(advert0_index, scale = 1)
advert0       = (advert0_prob >= u).astype(int)

np.mean(advert0)
```

```python
# Факт просмотра рекламной интеграции
advert = advert1 * block + advert0 * (1 - block)

# Доли людей, которые смотрят рекламу
print(np.mean(advert))
```

```python
# Рассмотрим корреляции
print(pd.DataFrame(data    = [np.corrcoef(advert, interest)[0, 1],
                              np.corrcoef(advert, block)[0, 1]],
                   index   = ['Corr(advert, interest)',
                              'Corr(advert, block)'],
                   columns = ['Оценка']))
```

**Генерация целевой переменной**


**Основная идея**
Наверно чтобы понять, какие можно построить зависимости, то нужно определять, что это за человек, который не смотрит рекламу. Мы полагаем, что рекламу смотрят менее раздраженные люди, которым все в целом интересно послушать. В моем мире это скорее взрослые люди без работы, а с детьми непонятно что.

1) **Если клиент посмотрел рекламу**

 В этом определеним смысле если человек смотрит рекламу, то он в целом смотрит видео не как дофаминовый наркоман, а просто хочет смотреть видео, так как ему очень интересно, тем самым, переменная создающая эндогенность должна иметь больший коэффициент при генерации количества часов  просмотра видео. По контрольным переменным, думаю можно предположить, что пусть немного сильнее влияют, чем у тех кто не посмотрел рекламу, так как он не смотрит видео на репите, и все характеристики начинают иметь большую роль

2) **Если клиент не посмотрел рекламу**

Тут я думаю сформированная нетерпимость к рекламе, отсюда перенасыщение просмотром видео,  то есть он смотрит просто чтобы смотреть, отсюда коэффициент при эндогенной переменной конечно положительный, но меньше чем в первом случае. Аналогичная логика и для контрольных переменных, так как это все уже имеет меньшее значение, если ты очень часто смотришь видео

```python
# Случайные ошибки
error0 = t.rvs(size = n, df = 15) * 8
error1 = expon.rvs(size = n, scale = 10) - 10

# Функция от контрольных переменных
# когда индивид посмотрел рекламу

g1_obs   = 140 * children / (2 - age - job) +age*job/3
g1_unobs =  0.7 * interest
g1       = g1_obs + g1_unobs


# когда индивид не посмотрел рекламу
g0_obs   = 120*children/ (5 - age - job) + age*job/4
g0_unobs =  0.5 * interest
g0       = g0_obs + g0_unobs



# Количество минут просмотра видео в зависимости от факта просмотра рекламы
minutes0 = g0 + error0
minutes1 = g1 + error1
```

Так как событие, когда пользователь, которые не посмотрел рекламу
смотрит видео дольше, чем пользователь, который просмотрел эту рекламу,
невозможно даже в тееории, то необходимо исключить такие случаи из наших сгенерированных данных данный случай. Приравняем тогда значения, количества просмотренных минут, и будем считать, что случай, когда пользователю не понравилось видео и он еще до показа рекламы выключил видео

```python
minutes1[minutes1 < minutes0] = minutes0[minutes1 < minutes0]

# Наблюдаемая зарплата
minutes = minutes1 * advert + minutes0 * (1 - advert)

# Для адекватности значений наблюдаемой переменной
minutes[minutes < 0] = 0
minutes[minutes > 720] = 720
```

```python
# Приблизительно оценим адекватность дисперсий
print(pd.DataFrame(data    = [np.var(error0),   np.var(g0),
                              np.var(g0_obs),   np.var(g0_unobs),
                              np.var(error1),   np.var(g1),
                              np.var(g1_obs),   np.var(g1_unobs)],
                   index   = ['Var(eps0)',     'Var(g0)',
                              'Var(g0_obs)',   'Var(g0_unobs)',
                              'Var(eps1)',     'Var(g1)',
                              'Var(g1_obs)',   'Var(g1_unobs)'],
                   columns = ['Оценка']))
```

```python
from sklearn.metrics import roc_auc_score

# предполагаемая прибыль
TN_prof = 0
TP_prof = 500
FP_prof = -1000
FN_prof = -170

thresholds = np.arange(0, 1, 0.1)
best_tresholds ={}
aucs = {}
# подбираем оптимальный порог для каждой модели
for model in [lr_best, rf_best, gb_best]:
  profits = []
  for t in thresholds:
    model.fit(X_train_class, y_train_class)
    y_preds = model.predict_proba(X_train_class)[:, 1]
    preds = (y_preds > t).astype(int)

    cfm = confusion_matrix(y_train_class, preds)
    TP, TN, FP, FN = cfm[1,1], cfm[0,0], cfm[0,1], cfm[1,0]

    profit = TN_prof * TN + TP_prof * TP + FP_prof * FP + FN_prof * FN
    profits.append(profit)
  best_tresholds[model.__class__.__name__] = thresholds[np.argmax(profits)]

# смотрим на прибыли на тестовой выборке
test_profits = {}

for model in [lr_best, rf_best, gb_best]:
    model.fit(X_train_class, y_train_class)
    y_preds = model.predict_proba(X_test_class)[:, 1]
    preds = (y_preds > best_tresholds[model.__class__.__name__]).astype(int)

    cfm = confusion_matrix(y_test_class, preds)
    TP, TN, FP, FN = cfm[1,1], cfm[0,0], cfm[0,1], cfm[1,0]

    profit = TN_prof * TN + TP_prof * TP + FP_prof * FP + FN_prof * FN
    test_profits[model.__class__.__name__] = profit
    aucs[model.__class__.__name__] = roc_auc_score(y_test_class, y_preds)
```

## 1. Математически запишите и содержательно проинтерпретируйте потенциальные исходы целевой переменной. Объясните, как они связаны с наблюдаемыми значениями целевой переменной.

Сформируем представления о поведении пользователя при просмотре видео в зависимости от факта просмотра рекламы.

Основная идея заключается в том, что интерес к видео, возраст, наличие детей и работы сильнее влияют на количество просматриваемого контента, когда индивид смотрит предпочитает смотреть рекламу, ведь, как уже упоминали выше, пользователь, которые спойкойно относятся к рекламе, не пресыщены общим количеством просмотренного контента и, таким образом, все факторы начинают иметь большую отдачу.

*Уравнение потенционального исхода целевой переменной при просмотре рекламы:*

$$\text{Minutes}_{1i} = \underbrace{\underbrace{0.7\times \text{Interest}_{i}}_{g_{1}^{\text{unobs}}} + \underbrace{14\times \frac{\text{Children}_{i}}{2 - \text{Age}_{i} - \text{Job}_{i}} + \frac{\text{Age}_{i} \times \text{Job}_{i}}{3}}_{g_{1}^{\text{obs}}}}_{g_{1}} + \varepsilon_{1i}\text{, где }\varepsilon_{1i}\sim \left(\text{EXP}(0.1) - 10\right)$$


*Уравнение потенционального исхода целевой переменной при отсутствии просмотра рекламы:*

$$\text{Minutes}_{0i} = \underbrace{\underbrace{0.5\times\text{Interest}_{i}}_{g_{0}^{\text{unobs}}} + \underbrace{12\times\frac{\text{Children}_{i}}{5 - \text{Age}_{i} - \text{Job}_{i}} + \frac{\text{Age}_{i} \times \text{Job}_{i}}{4}}_{g_{0}^{\text{obs}}}}_{g_{0}} + \varepsilon_{0i}\text{, где }\varepsilon_{0i}\sim \left(8\times t(15)\right)$$

Как известно, для одного и то же наблюдения одновременно не могут выполняться оба потенциональных исхода. Существует наблюдаемое значение целевой переменной. В данном случае наблюдаемое значения количества просмотренных минут $\text{Minutes}_{i}$ равно $\text{Minutes}_{0i}$, если не было просмотра рекламы, и $\text{Minutes}_{1i}$ в ином случае. Заметим, что ранее эти значения были так сгенерированы, что $\text{Minutes}_{1i} >= \text{Minutes}_{0i}$.

Таким образом, наблюдаемое значение целевой переменной можно записать следующим образом.

$$\text{Minutes}_{i} = \begin{cases}\text{Minutes}_{1i}\text{, если }\text{Advert}_{i}=1\\ \text{Minutes}_{0i}\text{, если }\text{Advert}_{i}=0\end{cases} = \\ =\text{Minutes}_{1i}\times\text{Advert}_{i} + \text{Minutes}_{0i}\times\left(1-\text{Advert}_{i}\right)$$

## 2. Используя симулированные вами, но недоступные в реальных данных потенциальные исходы (гипотетические значения), получите оценки среднего эффекта воздействия, условных средних эффектов воздействия и локального среднего эффекта воздействия. Результаты представьте в форме таблицы.

Настоящие эффекты воздействия:

$$\text{TE}_{i} = \text{Minutes}_{1i} - \text{Minutes}_{0i}$$

```python
# Настоящие эффекты воздействия (не наблюдаются в данных)
TE = minutes1 - minutes0
print(f'TE: {TE[:5]}')
```

Средний эффект воздействия:

$$\text{ATE} = \text{E}\left(\text{Minutes}_{1i} - \text{Minutes}_{0i}\right)$$

Если бы у нас были данные о потенциальных исходах $\text{Minutes}_{1i}$ и $\text{Minutes}_{0i}$, то мы могли бы точно оценить $\text{ATE}$ как:

$$\widehat{\text{ATE}} = \frac{1}{n}\sum\limits_{i=1}^{n}\text{Minutes}_{1i} - \text{Minutes}_{0i}$$

```python
# Точное приближение среднего эффекта воздействия, то есть
# с помощью оценки, недоступной с помощью реальных данных
ATE = np.mean(TE)
print(f'ATE = {ATE:.3f}')
```

Локальный средний эффект воздействия:

$$\text{LATE} = \text{E}(\text{Hours}_{1i} - \text{Hours}_{0i} | \text{Advert}_{1i} > \text{Advert}_{0i})$$

Здесь необходимо объяснить, кого в наших данных мы считаем за $\text{Always taker}$, $\text{Never taker}$, $\text{Complier}$ и $\text{Denier}$.

*   $\text{Always taker}$ - пользователь, который смотрит рекламу, независимо от того, у него есть блокировщик рекламы на устройстве или нет
*   $\text{Never taker}$ - пользователь, который не смотрит рекламу, независимо от того, у него есть блокировщик рекламы на устройстве или нет
*   $\text{Complier}$ - пользователь, который смотрит рекламу, если у него нет блокировщика рекламы, и не смотрит рекламу, если у него есть блокировщик
*   $\text{Denier}$ - пользователь, который смотрит рекламу, если у него есть блокировщика рекламы, и не смотрит рекламу, если у него нет блокировщика

Метрика $\text{LATE}$ считается для $\text{Compliers}$

```python
# Рассмотрим различные группы индивидов
ind_type = np.empty(n, dtype = 'U25')
ind_type[(advert1 == 1) & (advert1 == 1)] = 'Always taker'
ind_type[(advert1 == 0) & (advert1 == 0)] = 'Never taker'
ind_type[advert0 > advert1]               = 'Complier'
ind_type[advert0 < advert1]               = 'Denier'
```

## 3. Оцените средний эффект воздействия как разницу в средних по выборкам тех, кто получил и не получил воздействие. Опишите недостатки соответствующего подхода с учетом специфики рассматриваемой вами экономической проблемы.

Наивный подход предполагает оценивание $\text{ATE}$ как средней разницы в зарплатах людей с высшим образованием и без высшего образования.

$$\widehat{\text{ATE}}_{\text{naive}} = \frac{1}{n_{1}}\sum\limits_{i:\text{Advert}_{i}=1}\text{Minutes}_{1i} - \frac{1}{n_{0}}\sum\limits_{i:\text{Advert}_{i}=0}\text{Minutes}_{0i}$$

Где $n_{1}$ и $n_{0}$ - число людей с высшим образованием и без высшего образования соответственно.

```python
# Наивная оценка как разница в выборочных средних
ATE_naive = np.mean(minutes[advert == 1]) - np.mean(minutes[advert == 0])

# Сравнение точного приближения и наивной оценки
print(pd.DataFrame(data    = [ATE, ATE_naive],
                   index   = ['ATE', 'ATE naive'],
                   columns = ['Оценка']))
```

Как видим, оценка наивным образом все-таки отличается от настоящего ATE. Это свидетельствует о том, что не выполнено допущение о независимости:

$$\text{E}(\text{Minutes}_{1i}|\text{Advert}_{i}=1) = \text{E}(\text{Minutes}_{1i})\qquad \text{E}(\text{Minutes}_{0i}|\text{Advert}_{i}=0) = \text{E}(\text{Minutes}_{0i})$$

**а) Метод наименьших квадратов**

Для МНК оценики воспользуемся допущением об условной независимости:

$$\text{E}(\text{Minutes}_{1i}|\text{Advert}_{i}=1,X_{i}) = \text{E}(\text{Minutes}_{1i}|X_{i})\qquad \text{E}(\text{Minutes}_{0i}|\text{Advert}_{i}=0,X_{i}) = \text{E}(\text{Minutes}_{0i}|X_{i})$$

Попробуем оценить $\text{ATE}$, рассмотрев среднюю разницу в оценках минут, проведённых за просмотром видео, полученных с помощью МНК отдельно сперва по пользователям, которые посмотрели рекламу, а затем по пользователям, которые этого не сделали.

$$\widehat{\text{ATE}}_{\text{OLS}} = \frac{1}{n}\sum\limits_{i=1}^{n} \underbrace{\hat{\text{E}}\left(\text{Minutes}_{1i}|X_{i}\right) - \hat{\text{E}}\left(\text{Minutes}_{0i}|X_{i}\right)}_{\widehat{\text{CATE}}_{i}}$$

Где:


*   $\hat{\text{E}}\left(\text{Minutes}_{1i}|X_{i}\right)$ - оценка, полученная с использованием МНК оценок регрессионных коэффициентов $\beta$, полученных по выборке из пользователей, просмотревших рекламу $\text{Advert}_{i} = 1$.
*   $\hat{\text{E}}\left(\text{Minutes}_{0i}|X_{i}\right)$ - оценка, полученная с использованием МНК оценок регрессионных коэффициентов $\beta$, полученных по выборке из пользователей, не просмотревших рекламу $\text{Advert}_{i} = 0$.



```python
# Оценим средний эффект воздействия с помощью МНК

# МНК оценивание для пользователей, не посмотревших рекламу
y0 = df.loc[advert == 0, ['minutes']]
x0 = df.loc[advert == 0, ['job', 'children', 'age', 'interest']]
x0 = sm.add_constant(x0)
ols0 = sm.OLS(y0, x0).fit()

# МНК оценивание для пользователей, посмотревших рекламу
y1 = df.loc[advert == 1, ['minutes']]
x1 = df.loc[advert == 1, ['job', 'children', 'age', 'interest']]
x1 = sm.add_constant(x1)
ols1 = sm.OLS(y1, x1).fit()

# Оценим зарплаты при наличии и отсутствия высшего образования
# с помощью полученных МНК оценок
x = df.loc[:, ['job', 'children', 'age', 'interest']]
x = sm.add_constant(x)
  # МНК оценка E(wage0 | X) для всех индивидов
minutes0_ols = ols0.predict(x)
  # МНК оценка E(wage1 | X) для всех индивидов
minutes1_ols = ols1.predict(x)

# Оценки CATE
CATE_ols = np.array(minutes1_ols - minutes0_ols)

# Оценка ATE как средняя разница в прогнозах МНК оценок
ATE_ols = np.mean(CATE_ols)
```

**б) Условные математические ожидания**

Оценки условных математических ожиданий можно получить двумя разными способами: S-learner -
оценка $\hat{\text{E}}(\text{Minutes}_{i}|\text{Advert}_{i},X_{i})$ по всей выборке, или T-learner - оценка
отдельно $\hat{\text{E}}(\text{Minutes}_{i}|\text{Advert}_{i}=1,X_{i})$ и $\hat{\text{E}}(\text{Minutes}_{i}|\text{Advert}_{i}=0,X_{i})$ по группе воздействия и по контрольной
группе соответственно. Заметим, что для обоих способов снова требуется предположения об условной независимости. Воспользуемся обоими способами и затем сравним результаты

Где:

*   $\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i},\text{Advert}_{i}=1\right)$ - оценка, полученная с использованием метода машинного обучения по выборке из пользователей, просмотревших рекламу $\text{Advert}_{i} = 1$.
*   $\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i},\text{Advert}_{i}=0\right)$ - оценка, полученная с использованием метода машинного обучения по выборке из пользователей, не просмотревших рекламу $\text{Advert}_{i} = 0$.

*S-learner*

Оценка такая же, как в T-learner. Отличаются только методы оценки условных математических ожиданий.

$$\widehat{\text{ATE}}^{\text{S-learner}} = \frac{1}{n}\sum\limits_{i=1}^{n} \hat{\text{E}}\left(\text{Minutes}_{i}|X_{i},\text{Advert}_{i}=1\right) - \hat{\text{E}}\left(\text{Minutes}_{i}|X_{i},\text{Advert}_{i}=0\right)$$

Где:

*   $\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i},\text{Advert}_{i}=1\right)$ - оценка, полученная с использованием метода машинного обучения по всей выборке.
*   $\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i},\text{Advert}_{i}=0\right)$ - оценка, полученная с использованием метода машинного обучения по всей выборке.

```python
# Оценивание ATE и CATE с помощью S-learner

# Подготовим данные
y = df.loc[:, ['minutes']]
x = df.loc[:, ['job', 'children', 'age', 'interest','advert']]

# Используем лучшую модель, которую нашли ранее
cat_reg_best_2 = CatBoostRegressor(depth = compare_cat_reg_df['best']['depth'],
                                 learning_rate = compare_cat_reg_df['best']['learning_rate'],
                                 iterations = compare_cat_reg_df['best']['iterations'],
                                 verbose=False)
cat_reg_best_2.fit(x, y)

# Оценки E(Minutes | X, Advert = 0) для всех индивидов.
x0 = deepcopy(x)
x0["advert"] = 0
minutes0_rf2 = cat_reg_best_2.predict(x0)

# Оценки E(Minutes | X, Advert = 1) для всех индивидов.
x1 = deepcopy(x)
x1["advert"] = 1
minutes1_rf2 = cat_reg_best_2.predict(x1)

# Оценки CATE
CATE_S = minutes1_rf2 - minutes0_rf2

# Оценка ATE
ATE_S = np.mean(CATE_S)
```

**в) Взвешивания на обратные вероятности**

Метод взвешивания на обратные вероятности можно использовать, когда выполнена предпосылка об условной независимости. Оценка, получаемая с помощью взвешивания на обратные вероятности $\text{IPW}$, имеет вид:

$$\widehat{\text{ATE}}^{\text{IPW}} = \frac{1}{n}\sum\limits_{i=1}^{n}\frac{\text{Advert}_{i}\times\text{Minutes}_{i}}{\hat{P}\left(\text{Advert}_{i}=1|X_{i}\right)} - \frac{\left(1 - \text{Advert}_{i}\right)\times\text{Minutes}_{i}}{1 - \hat{P}\left(\text{Advert}_{i}=1|X_{i}\right)}$$

Где условные вероятности $\hat{P}\left(\text{Advert}_{i}=1|X_{i}\right)$ оценивается с помощью методов классификации. В данном случае воспользуемся градиентным бустингом

```python
# Оценивание с помощью обратного взвешивания на вероятности IPW

# Подготовим данные
target = df.loc[:, ['advert']]
features = df.loc[:, ['job', 'children', 'age', 'interest']]

# Подготовим метод машинного обучения
cbc = CatBoostClassifier(**best_params_cat, verbose=False)
cbc.fit(features, target)

# Оценим условные вероятности P(educ = 1 | X)
prob_cbc = cbc.predict_proba(features)[:, 1]

# Оценим псевдоисходы
minutes_pseudo = (minutes * advert) / prob_cbc - (minutes * (1 - advert)) / (1 - prob_cbc)

# Оценим ATE
ATE_IPW = np.mean(minutes_pseudo)
```

**г) Метод, обладающий двойной устойчивостью**

Данный метод объединяет методы взвешивания на обратные вероятности метод оценки через условные математические ожидания, проблемой которых являлось то,  точность оценок каждого из этих способов зависит от точности оценок соответствующих условных математических ожиданий или вероятностей. Если они оценены неточно, то и итоговая оценка ATE также будет неточной. Поэтому метод, обладающий двойной устойчивостью, совмещает оба способа, чтобы оценка ATE
оказывалась состоятельной, если по крайней мере один из них дает состоятельную оценку

Методом с двойной устойчивостью средний эффект воздействия оценивается как:

$$\widehat{\text{ATE}}^{\text{DR}} = \frac{1}{n}\sum\limits_{i=1}^{n}\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i}, T_{i}=1\right) - \hat{\text{E}}\left(\text{Minutes}_{i}|X_{i}, T_{i}=0\right) + \frac{\text{Advert}_{i}\times\left(\text{Minutes}_{i}-\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i}, T_{i}=1\right)\right)}{\hat{P}(\text{Advert}_{i}=1|X_{i})} - \frac{\left(1-\text{Advert}_{i}\right)\times\left(\text{Minutes}_{i}-\hat{\text{E}}\left(\text{Minutes}_{i}|X_{i}, T_{i}=0\right)\right)}{1 - \hat{P}(\text{Advert}_{i}=1|X_{i})}$$

```python
# Оценим ATE с помощью DR метода с двойной устойчивостью
ATE_DR = np.mean(minutes1_rf1 - minutes0_rf1 + advert * (minutes - minutes1_rf1) / prob_cbc - (1 - advert) * (minutes - minutes0_rf1) / (1 - prob_cbc))
```

**Вывод:** Оценка методом, обладающим двойной устойчивости, обладает достойным результатом

**Метод трансформации классов**

Методом трансформации классов условные эффекты воздействия оцениваются как:
$$\widehat{\text{CATE}}_{i}^{\text{IPW}}=\hat{\text{E}}(\widehat{\text{Wage}}_{i}^{*}|X_{i})$$

Где $\hat{\text{E}}(\widehat{\text{Minutes}}_{i}^{*}|X_{i})$ это оценка условного математического ожидания оценки псевдоисхода, определяемой как:

$$\widehat{\text{Wage}}_{i}^{*} = \frac{\text{Advert}_{i}\times\text{Minutes}_{i}}{\hat{P}\left(\text{Advert}_{i}=1|X_{i}\right)} - \frac{\left(1 - \text{Advert}_{i}\right)\times\text{Minutes}_{i}}{1 - \hat{P}\left(\text{Advert}_{i}=1|X_{i}\right)}$$



```python
# Оценим CATE методом трансформации классов

# Метод прогнозирования условных математических ожиданий оценок псевдоисходов
rf3 = CatBoostRegressor(depth = compare_cat_reg_df['best']['depth'],
                        learning_rate = compare_cat_reg_df['best']['learning_rate'],
                        iterations = compare_cat_reg_df['best']['iterations'],
                        verbose=False)

rf3.fit(features, minutes_pseudo)

# Оценки CATE (метод трансформации классов)
CATE_CT = rf3.predict(features)
```

```python
print(f'Метод трансформации классов: {CATE_CT[:5]}')
```

```python
# Метод оценивания
rf4 = CatBoostRegressor(depth = compare_cat_reg_df['best']['depth'],
                        learning_rate = compare_cat_reg_df['best']['learning_rate'],
                        iterations = compare_cat_reg_df['best']['iterations'],
                        verbose=False)

# Оценим средний эффект воздействия с помощью МНК

# МНК оценивание для пользователей, не посмотревших рекламу
y0 = df.loc[advert == 0, ['minutes']]
x0 = df.loc[advert == 0, ['job', 'children', 'age', 'interest']]

# МНК оценивание для пользователей, посмотревших рекламу
y1 = df.loc[advert == 1, ['minutes']]
x1 = df.loc[advert == 1, ['job', 'children', 'age', 'interest']]

# Обучение оценивать E(Minutes | X, Advert = 0)
rf4.fit(x0, y0)

# Оценка целевой переменной при отсутствии воздействия в тестовой группе.
y0_pred = rf4.predict(x1)

# Обучение оценивать E(Minutes | X, Advert = 1)
rf4.fit(x1, y1)

# Оценка целевой переменной при наличии воздействия в контрольной группе.
y1_pred = rf4.predict(x0)

# Введем вспомогательные оценки
d1 = y1['minutes'] - y0_pred
d0 = y1_pred - y0['minutes']

# Метод оценивания
rf5 = CatBoostRegressor(depth = compare_cat_reg_df['best']['depth'],
                        learning_rate = compare_cat_reg_df['best']['learning_rate'],
                        iterations = compare_cat_reg_df['best']['iterations'],
                        verbose=False)

# Обучение оценивать E(^D0 | X)
rf5.fit(x0, d0)

# Оценка для E(^D0)
d0_pred = rf5.predict(features)

# Обучение оценивать E(^D1 | X)
rf5.fit(x1, d1)

# Оценка для E(^D1)
d1_pred = rf5.predict(features)

# Значения CATE
CATE_X = (1  - prob_cbc) * d1_pred + prob_cbc * d0_pred
```
